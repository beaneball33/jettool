```python
#以下四行是用來控制範例可以找到jettool
import os,sys,inspect
currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir) 
#import此功能套件
import jettool.dataset.dbapi as dbapi
```

```python
# 取得該使用者能存取的所有table的資訊
dbapi.get_info("CEwRaqW61gOe5KcIssqCqIleChUlSu")
```

```python
# 查詢所有的國別的資料庫資訊 
dbapi.get_market("CEwRaqW61gOe5KcIssqCqIleChUlSu")
```

```python
# 查詢所有資料分類的資料庫 
dbapi.get_category("CEwRaqW61gOe5KcIssqCqIleChUlSu")
```

```python

# 查詢按國別分類的資料表完整清單   
# 查詢所有資料表資訊內容
table_list = dbapi.get_tables("CEwRaqW61gOe5KcIssqCqIleChUlSu")
table_list
```

```python
# 根據索引目錄的資料表清單構造，回傳 dataframe方便檢視
dbapi.get_tables_info("CEwRaqW61gOe5KcIssqCqIleChUlSu",market='TWN')
```

```python
# 根據已知的table名稱，查詢 mapping到別的國家的資料表清單
dbapi.get_table_mapping(api_key="CEwRaqW61gOe5KcIssqCqIleChUlSu",table_name='TWN/AIND')
```

```python
# 使用tejapi.search_table進行交集或聯集查詢
# 輸入要搜尋的關鍵字，並且產出完整會計科目的中文名稱     
dbapi.search_column(api_key="CEwRaqW61gOe5KcIssqCqIleChUlSu",market='TWN',keyword='報酬率 月',condition='and')
```
